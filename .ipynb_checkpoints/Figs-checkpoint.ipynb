```python
import pathlib
import subprocess
import h5py
import numpy as np
import matplotlib.pyplot as plt
import dedalus.public as d3

print(subprocess.check_output("find analysis | sort", shell=True).decode())
```

```python
# function to plot temperature fields
def temp_field_SS(profile, ind):
    # create a figure and axis
    fig, ax = plt.subplots(figsize = (10,3))
    
    # plot temperature field 
    im = ax.imshow(profile[-1, :, :].T, origin='lower',extent=[0, 4, -0.5, 0.5], aspect='auto', cmap='coolwarm')

    # add colorbar
    cbar = fig.colorbar(im, ax=ax, orientation='vertical')
    
    # set ticks and tick labels
    cbar.set_ticks(np.arange(0, 1.1, 0.2))
    cbar.set_ticklabels([f'{val:.1f}' for val in np.arange(0, 1.1, 0.2)])
    cbar.set_label('Temperature')

    # set labels and title
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Z-axis')
    ax.set_title(f'Temperature Field for Ra = 1e{ind}')

    # save the plot
    plt.savefig(f'T_field_Ra1e{ind}')
    
```

```python
# Load data and store temperature profiles and Nusselt numbers
averaged_temperature_profiles = []

# load data and store temperature fields
for j in range(3,5):
    with h5py.File(f"analysisRa2e{j}/analysis_s1.h5", mode='r') as file:
        # load datasets
        T = file['tasks']['b']

        # extract the temperature data for all x values
        temperature_profiles = T[:, :, :]

        # compute the mean along the x-axis 
        averaged_temperature_profile = np.mean(temperature_profiles, axis=1)
        averaged_temperature_profiles.append(averaged_temperature_profile)

        temp_field_SS(temperature_profiles, j)
        
for j in range(5,7):
    with h5py.File(f"analysisRa1e{j}/analysis_s2.h5", mode='r') as file:
        # load datasets
        T = file['tasks']['b']

        # extract the temperature data for all x values
        temperature_profiles = T[:, :, :]

        # compute the mean along the x-axis 
        averaged_temperature_profile = np.mean(temperature_profiles, axis=1)
        averaged_temperature_profiles.append(averaged_temperature_profile)
    
        temp_field_SS(temperature_profiles, j)


with h5py.File(f"analysisRa1e7/analysis_s3.h5", mode='r') as file:
    # load datasets
    T = file['tasks']['b']

    # extract the temperature data for all x values
    temperature_profiles = T[:, :, :]

    # compute the mean along the x-axis 
    averaged_temperature_profile = np.mean(temperature_profiles, axis=1)
    averaged_temperature_profiles.append(averaged_temperature_profile)
    
    temp_field_SS(temperature_profiles, 7)

with h5py.File(f"analysisRa1e8/analysis_s4.h5", mode='r') as file:
    # load datasets
    T = file['tasks']['b']

    # extract the temperature data for all x values
    temperature_profiles = T[:, :, :]

    # compute the mean along the x-axis 
    averaged_temperature_profile = np.mean(temperature_profiles, axis=1)
    averaged_temperature_profiles.append(averaged_temperature_profile)
    
    temp_field_SS(temperature_profiles, 8)
```

```python
fig = plt.figure(figsize = (8,5))
x = np.linspace(0,1.0)
z = np.linspace(-0.5,0.5,64)
conduction = 0.5 - x
plt.plot(x,conduction, '--k', label = 'conduction')

for i in range(len(averaged_temperature_profiles)):
    # Plot temperature against the z-axis
    plt.plot(averaged_temperature_profiles[i][-1, :], z, label = f'Ra = 1e{i+3}')  # plotting last write

plt.xlabel('Horizontally Averaged Temperature')
plt.ylabel('Z-axis')
#plt.title(f'Horizontally Averaged Temperature Profiles')
plt.grid(ls = '--')
plt.legend()
plt.savefig('temp_profiles')
plt.show()
```

```python
# load all data this time but only to extract nusselt numbers
Nu_bottom = []
Nu_top = []
Nu_errs = []

for j in range(1,7):
    with h5py.File(f"data/analysis{j}/analysis_s1.h5", mode='r') as file:
        Nu = file['tasks']['Nusselt Number']
        #print(Nu)
        Nu_t = Nu[-1,0,-1]  # last value corresponds to the top
        Nu_b = Nu[-1,0,0]   # first value corresponds to the bottom
 
        # save nusselt numbers 
        Nu_bottom.append(Nu_b)
        Nu_top.append(Nu_t)
        Nu_errs.append(np.abs(Nu_t-Nu_b))
        print(f'Number {j}: Nu_t=',Nu_t,'Nu_b=',Nu_b)

for j in range(7,12):
    with h5py.File(f"data/analysis{j}/analysis_s2.h5", mode='r') as file:
        Nu = file['tasks']['Nusselt Number']
        #print(Nu)
        Nu_t = Nu[-1,0,-1]  # last value corresponds to the top
        Nu_b = Nu[-1,0,0]   # first value corresponds to the bottom
 
        # save nusselt numbers 
        Nu_bottom.append(Nu_b)
        Nu_top.append(Nu_t)
        Nu_errs.append(np.abs(Nu_t-Nu_b))
        print(f'Number {j}: Nu_t=',Nu_t,'Nu_b=',Nu_b)

for j in range(12,16):
    with h5py.File(f"data/analysis{j}/analysis_s3.h5", mode='r') as file:
        Nu = file['tasks']['Nusselt Number']
        #print(Nu)
        Nu_t = Nu[-1,0,-1]  # last value corresponds to the top
        Nu_b = Nu[-1,0,0]   # first value corresponds to the bottom
 
        # save nusselt numbers 
        Nu_bottom.append(Nu_b)
        Nu_top.append(Nu_t)
        Nu_errs.append(np.abs(Nu_t-Nu_b))
        print(f'Number {j}: Nu_t=',Nu_t,'Nu_b=',Nu_b)


with h5py.File(f"data/analysis16/analysis_s4.h5", mode='r') as file:
    Nu = file['tasks']['Nusselt Number']
    #print(Nu)
    Nu_t = Nu[-1,0,-1]  # last value corresponds to the top
    Nu_b = Nu[-1,0,0]   # first value corresponds to the bottom
 
    # save nusselt numbers 
    Nu_bottom.append(Nu_b)
    Nu_top.append(Nu_t)
    Nu_errs.append(np.abs(Nu_t-Nu_b))
    print('Number 16: Nu_t=',Nu_t,'Nu_b=',Nu_b)
```

```python
# average nusselt number for plotting
Nu_avg = np.zeros(len(Nu_bottom))
for i in range(len(Nu_bottom)):
    Nu_avg[i] = (Nu_bottom[i] + Nu_top[i]) / 2

print(Nu_avg)
```

```python
# all the Rayleigh numbers used for simulation
Ras = [1e3,2e3,5e3,1e4,2e4,5e4,1e5,2e5,5e5,1e6,2e6,5e6,7e6,1e7,2e7,2e8]

fig = plt.figure(figsize = (8,4))
# plotting error for Nusselt number as a function of Ra
plt.plot(Ras,Nu_errs, ':ok')
plt.ylabel('Difference |Nu$_{t}$ - Nu$_{b}$|')
plt.xlabel('Rayleigh Number, Ra')
plt.xscale('log')
plt.grid(ls = '--')
plt.savefig('NuerrvsRa')
```

```python
empirical_scaling = 0.138*np.array(Ras)**(0.285)  # johnston and doering 2009

fig = plt.figure(figsize=(8,4))

# plotting Nusselt number as a function of Ra and empirical scaling to check for consistency
plt.plot(Ras,empirical_scaling, ':r',label = '$0.138Ra^{0.285}$ Johnston and Doering 2009')
plt.plot(Ras,Nu_avg, 'ok', label = 'simulation data')
plt.ylabel('Nusselt Number, Nu')
plt.xlabel('Rayleigh Number, Ra')
plt.xscale('log')
plt.yscale('log')
plt.grid(ls = '--')
plt.legend()
plt.savefig('NuvsRa')
```
